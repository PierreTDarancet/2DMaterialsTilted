# Tilted Orthorhombic Supercell Generator

This notebook generates specific low-symmetry supercells for 2D materials (MoS$_2$) under periodic boundary conditions. 

It solves the geometric problem of fitting a tilted material sheet (diagonal) inside an orthorhombic box while preserving a fixed perpendicular vacuum spacing $h$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pymatgen.core import Structure, Lattice

# --- 1. CRYSTAL GEOMETRY ---

def get_correct_ortho_mos2(a_lattice=3.18, thickness=3.12, vacuum=20.0):
    """
    Generates the CORRECT 6-atom Rectangular MoS2 cell (2H phase).
    Geometry: Centered Rectangular
    a_rect = a
    b_rect = a * sqrt(3)
    """
    b_lattice = a_lattice * np.sqrt(3)
    lattice = Lattice.from_parameters(a_lattice, b_lattice, vacuum, 90, 90, 90)
    
    # Fractional z-coordinates
    z_mo = 0.5
    z_delta = (thickness / 2.0) / vacuum
    z_s_top = 0.5 + z_delta
    z_s_bot = 0.5 - z_delta
    
    species = ["Mo", "Mo", "S", "S", "S", "S"]
    
    # Standard 2H coordination (S at 1/3 offset)
    coords = [
        [0.0, 0.0, z_mo],           # Mo1
        [0.5, 0.5, z_mo],           # Mo2
        [0.0, 1/3, z_s_bot],        # S1 Bot
        [0.0, 1/3, z_s_top],        # S1 Top
        [0.5, 0.5 + 1/3, z_s_bot],  # S2 Bot
        [0.5, 0.5 + 1/3, z_s_top],  # S2 Top
    ]
    
    coords = [[c[0]%1.0, c[1]%1.0, c[2]] for c in coords]
    return Structure(lattice, species, coords)

# --- 2. GEOMETRY SOLVER ---

def solve_box_geometry(length_sheet, h_spacing):
    """Calculates new box dimensions and tilt angle theta."""
    if length_sheet < 2 * h_spacing:
        return None, None, None
        
    term = np.sqrt(length_sheet**4 - 4 * (length_sheet * h_spacing)**2)
    L_planar = np.sqrt((length_sheet**2 + term) / 2)
    L_z = (length_sheet * h_spacing) / L_planar
    theta = np.arctan(L_z / L_planar)
    return L_planar, L_z, theta

def generate_tilted_cell(target_h, rotation_axis='b', a_param=3.18):
    # 1. Get Base Unit Cell
    struct = get_correct_ortho_mos2(a_lattice=a_param)
    a_rect = struct.lattice.a
    b_rect = struct.lattice.b
    
    # 2. Determine Supercell Size (N)
    unit_length = b_rect if rotation_axis == 'a' else a_rect
    min_N = int(np.ceil((2.01 * target_h) / unit_length))
    N = max(min_N, 1)
    
    print(f"Target h={target_h}A. Auto-selected Supercell N={N}.")

    # 3. Create Strip Supercell
    if rotation_axis == 'b':
        struct.make_supercell([N, 1, 1])
        total_length = N * a_rect
    else:
        struct.make_supercell([1, N, 1])
        total_length = N * b_rect

    # 4. Solve Geometry
    L_planar, L_z, theta = solve_box_geometry(total_length, target_h)
    
    # 5. Rigid Body Rotation
    cart_coords = struct.cart_coords
    z_mean = np.mean(cart_coords[:, 2])
    cart_coords[:, 2] -= z_mean

    if rotation_axis == 'b':
        # Rotate around Y (Cartesian), X -> Z
        c, s = np.cos(theta), np.sin(theta)
        R = np.array([[c, 0, -s], [0, 1, 0], [s, 0, c]])
        new_lattice = Lattice.from_parameters(L_planar, b_rect, L_z, 90, 90, 90)
    else:
        # Rotate around X (Cartesian), Y -> Z
        c, s = np.cos(theta), np.sin(theta)
        R = np.array([[1, 0, 0], [0, c, -s], [0, s, c]])
        new_lattice = Lattice.from_parameters(a_rect, L_planar, L_z, 90, 90, 90)

    rotated_coords = np.dot(cart_coords, R.T)
    
    new_struct = Structure(new_lattice, struct.species, rotated_coords, coords_are_cartesian=True)
    return new_struct, theta, N
```

### Example 1: Rotation around B-axis
This tilts the short edge ($a$) into the $z$-direction.

```python
struct_b, theta_b, N_b = generate_tilted_cell(target_h=15.0, rotation_axis='b')

print(f"Generated structure with {len(struct_b)} atoms.")
print(f"Tilt Angle: {np.degrees(theta_b):.2f} degrees")
print(f"Lattice: {struct_b.lattice.abc}")

# Write to file
struct_b.to(fmt="poscar", filename="POSCAR_tilt_b.vasp")
```

### Visualization Helper
A quick 3D plot to verify the atoms are forming a diagonal plane.

```python
def plot_structure(struct, title="Structure"):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    # Extract coords
    coords = struct.cart_coords
    x, y, z = coords[:, 0], coords[:, 1], coords[:, 2]
    
    # Color by species
    colors = ['purple' if str(s) == 'Mo' else 'yellow' for s in struct.species]
    
    # Plot atoms
    ax.scatter(x, y, z, c=colors, s=100, edgecolors='k', alpha=0.8)
    
    # Plot Box Boundaries
    a, b, c = struct.lattice.abc
    # Draw simple bounding box lines
    for i in [0, a]:
        for j in [0, b]:
            ax.plot([i, i], [j, j], [0, c], color='k', linestyle=':', alpha=0.3)
    for i in [0, a]:
        for k in [0, c]:
            ax.plot([i, i], [0, b], [k, k], color='k', linestyle=':', alpha=0.3)
    for j in [0, b]:
        for k in [0, c]:
            ax.plot([0, a], [j, j], [k, k], color='k', linestyle=':', alpha=0.3)

    ax.set_xlabel('X (A)')
    ax.set_ylabel('Y (A)')
    ax.set_zlabel('Z (A)')
    ax.set_title(title)
    
    # Set aspect ratio equal for realistic viewing
    ax.set_box_aspect([a, b, c])
    
    plt.show()

plot_structure(struct_b, title="Tilt around B-axis")
```
